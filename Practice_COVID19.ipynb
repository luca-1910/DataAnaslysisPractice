# Re-arrange and clean dataset for readability

```python
df= df.drop(columns=["Unnamed: 0", "Unnamed: 1"])
```

```python
df.head(14)
```

```python
df= df.drop(columns=["Unnamed: 3"])
```

```python
df= df.drop(columns=["2022/2023"])
```

```python
df.columns = ["Code", "Description", "Collection Frequency", "Q1", "Q2", "Q3", "Q4"]
```

```python
df.head(14)
```

```python
df["Code"] = df["Code"].fillna(method="ffill")
```

```python
df.head(14)
```

```python
df.replace("NC", np.nan, inplace=True)
```

```python
df.dtypes
```

```python
df.head(14)
```

# Apply interpolation to fill empty values for data visualisation purposes 

```python
#Ensure quaterly columns are numeric
df[["Q1","Q2","Q3","Q4",]]=df[["Q1","Q2","Q3","Q4",]].astype(float)
```

```python
#Interpolate missing values across quaters(row-wise)
df[["Q1","Q2","Q3","Q4",]]=df[["Q1","Q2","Q3","Q4",]].interpolate(axis=1)
```

```python
df.head(14)
```

# Line plot: Comparing Q1-Q4 cases by Code=[COVID19]

```python
sns.set(style="whitegrid")
```

```python
covid19 = df[df["Code"]=="COVID19"]
plt.plot(["Q1","Q2","Q3","Q4",], covid19[["Q1","Q2","Q3","Q4",]].values.flatten(), marker='o')
plt.title=("COVID19 Quaterly Trend")
plt.ylabel("Cases")
plt.xlabel("Quarter")
plt.show()
```
